```python
import pandas as pd
```

```python
migrant_df = ('34170do001_migrants_census_2016_datacube_2_victoria.xls')
read_migrant = pd.ExcelFile(migrant_df)
print(read_migrant.sheet_names)
```

```python
migrant_7 = read_migrant.parse('Table 7')
migrant_7.head(5)
```

```python
migrant_7_clean = migrant_7.dropna()
migrant_7_clean.head(5)
```

```python
migrant_7_clean = migrant_7_clean.rename({'Australian Bureau of Statistics': 'Country', 
                                          'Unnamed: 1': 'Speaks English only',
                                          'Unnamed: 2': 'Very well',
                                          'Unnamed: 3': 'Well',
                                          'Unnamed: 4': 'Not well',
                                          'Unnamed: 5': 'Not at all',
                                          'Unnamed: 6': 'Not stated',
                                          'Unnamed: 7': 'Total'}, axis=1)
migrant_7_clean.head(5)
```

```python
migrant_7_clean.columns.values[0] = "Country_of_Birth"
migrant_7_clean.head(5)
```

```python
migrant_clean = migrant_7_clean.reset_index()
```

```python
migrant_clean = migrant_clean.reset_index(drop=True)
# migrant_clean
```

```python
migrant_full = migrant_clean.drop(['index'],axis=1)
migrant_full.head()
```
